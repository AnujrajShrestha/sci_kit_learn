```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error
import numpy as np

data=pd.read_csv("sheet1.csv")
df=pd.DataFrame(data)
df=df.dropna()
df
```

```python
X=data[['Hours']]
y=data['Score']

model=LinearRegression()
model.fit(X,y)

predictedScore=model.predict(X)

mae=mean_absolute_error(y,predictedScore)
mse=mean_squared_error(y,predictedScore)
rmse=np.sqrt(mse)

print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")

newPrediction=float(input("Enter hours: "))
newPred=model.predict([[newPrediction]])
print(f"Prediction for {newPrediction} is score= {newPred}")

```
