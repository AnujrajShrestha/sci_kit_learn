```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('Employee.csv')
df=pd.DataFrame(data)
df
```

```python
standardScaler=StandardScaler()
standardScaled=standardScaler.fit_transform(df[['JoiningYear','Age']])

pd.DataFrame(standardScaled,columns=['JoiningYear','Age'])
```

```python
minMaxScaler=MinMaxScaler()
minMaxScaled=minMaxScaler.fit_transform(df[['JoiningYear','Age']])

pd.DataFrame(minMaxScaled,columns=['JoiningYear','Age'])
```

```python
X=df[['JoiningYear']]
y=df[['Age']]

XTrain,XTest,yTrain,yTest=train_test_split(X,y,test_size=0.5,random_state=42)

print("Trainin data and Testind data of X 'JoiningYear'")
print(XTrain,XTest)
print("Trainin data and Testind data of y 'Age'")
print(yTrain,yTest)
```
